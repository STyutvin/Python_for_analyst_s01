## Задача N1

На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе.

```python
# Зададим функцию по суммированию всех фруктов
# Теоретически данная задача подразумевает использование (**kwargs), но я хочу взять для этого простой словарь (и никто мне не запретит)
# Тем более это не противоречит условию задачи

def store(fruits: dict):
    total_fruits=sum (fruits.values())
    return total_fruits
```

```python
# Проверим её работоспособность, задав какую-нибудь переменную

some_fruits = {'banana': 5, 'mango': 7, 'apple': 8, 'oranges':9}
store(some_fruits)
```

## Задача N2

Дан список с затратами на рекламу. Но в данных есть ошибки, некоторые затраты имеют отрицательную величину. Удалите такие значения из списка и посчитайте суммарные затраты
[100, 125, -90, 345, 655, -1, 0, 200].

```python
# Зададим указанный в условии массив в качестве списка

advertising_costs=[100, 125, -90, 345, 655, -1, 0, 200]
```

```python
# Уберем отрицательные значения

new_costs=[i if i>0 else 0 for i in advertising_costs]
new_costs
```

```python
# Теперь найдем суммарные затраты

all_costs=sum(i for i in new_costs)
all_costs
```

Суммы покупок по датам
[1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]
1.   Найдите, какая выручка у компании в ноябре (используйте list comprehensions).
2.   Найдите выручку компании в зависимости от месяца.
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка (спользуйте аннотирование типов.).

*Вот честно, не помню я, чтобы на лекиях по Питону давали задачи на обработку дат и времени именно в таком виде. По всей видимости придется "гуглить". Кстати, в таких случаях вспоминается бородатая байка про одного джентльмена, который в своем резюме написал: "При наличии Гугла богоподобен".*

```python
# Выполним первый пункт и посчитаем выручку компании в ноябре

date = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29',
        '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02',
        '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09',
        '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
money = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592,
         2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]
 
print(sum([m for d, m in zip(date, money) if d[5:7] == "11"]))
```

```python
# А теперь вторая часть задачи, где напишем функцию для выручки по месяцам
# Где ключи - это месяцы, а значения - это выручка

def sum_in_month(date: list, money: list) -> dict:
    res = {}
    for d, m in zip(date, money):
        if d[5:7] in res.keys():
            res[d[5:7]] = res[d[5:7]] + m
        else:
            res[d[5:7]] = m
    return res
 
print(sum_in_month(date, money))
```
